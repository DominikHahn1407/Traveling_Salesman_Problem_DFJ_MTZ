```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
def create_statistics_df(json_files):
    dfj_times = []
    mtz_times = []

    dfj_size_1 = []
    mtz_size_1 = []
    dfj_size_2 = []
    mtz_size_2 = []

    # each file contains one execution
    for file in json_files:
        with open(file, "r") as f:
            execution = json.load(f)

        dfj_times.append(execution["dfj_time_ms"])
        mtz_times.append(execution["mtz_time_ms"])

        dfj_size_1.append(execution["dfj_size_bytes_1"])
        mtz_size_1.append(execution["mtz_size_bytes_1"])

        dfj_size_2.append(execution["dfj_size_bytes_2"])
        mtz_size_2.append(execution["mtz_size_bytes_2"])

    rows = []

    for algorithm, times, size_1, size_2 in [
        ("DFJ", dfj_times, dfj_size_1, dfj_size_2),
        ("MTZ", mtz_times, mtz_size_1, mtz_size_2)
    ]:
        time_series = pd.Series(times)
        size_1_series = pd.Series(size_1)
        size_2_series = pd.Series(size_2)

        avg_time = time_series.mean()
        std_time = time_series.std()

        cv = (std_time / avg_time) * 100 if avg_time != 0 else None

        rows.append({
            "algorithm": algorithm,
            "average_time_ms": round(avg_time, 2),
            "std_time_ms": round(std_time, 2),
            "coefficient_of_variation_percent": round(cv, 2),
            "average_size_kb": round(size_1_series.mean() / 1000, 2),
            "std_size_kb": round(size_1_series.std() / 1000, 2),
            "number_of_runs": len(times)
        })

    stats_df = pd.DataFrame(rows)

    return stats_df
```

## Western Sahara

```python
data_dir = Path("results")

json_data = {}

for file in data_dir.rglob("western_sahara*.json"):
    with open(file, "r", encoding="utf-8") as f:
        json_data[str(file)] = json.load(f)

create_statistics_df(json_data)
```
